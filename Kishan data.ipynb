### 5. What is the average, minimum and maximum age group for people working in different sectors


```python
avg_Age_differentsectors = pd.DataFrame(df.groupby(['workclass']).mean()['Age'], columns = ['Age'])
avg_Age_differentsectors
```

```python
Min_Age_differentsectors = pd.DataFrame(df.groupby(['workclass']).min()['Age'], columns = ['Age'])
Min_Age_differentsectors
```

```python
Max_Age_differentsectors = pd.DataFrame(df.groupby(['workclass']).max()['Age'], columns = ['Age'])
Max_Age_differentsectors
```
